```python
def funcao_fibonacci(n):
  fibonacci = [0,1]

  for _ in range(n):
    fibonacci.append(fibonacci[-1] + fibonacci[-2])
  return fibonacci

num_termos = int(input('Digite o número: '))

resultado = funcao_fibonacci(num_termos)
print(resultado)
```

```python
numero = int(input('Digite um número: '))

def somando_numeros(numero):
    somando_numeros_local = 0
    while numero > 0:
        somando_numeros_local += numero % 10
        numero //= 10
    return somando_numeros_local

print(f'A soma dos dígitos é: {somando_numeros(numero)}')
```

```python
n = int(input('Digite um número: '))

def fatorial(n):
  if n == 0:
    return 1

  elif n < 0:
    return 'Número inválido'
  else:

    return n * fatorial(n-1)

resultado_fatorial = fatorial(n)

print(f'O fatorial de {n} é: {resultado_fatorial}')
```

```python
palíndromo = str(input('Digite uma palavra: '))

def analisar_palíndromo(palavra):
  if palavra == palavra[::-1]:
    return True
  else:
    return False

resultado = analisar_palíndromo(palíndromo)

if resultado == True:
  print(f'A palavra é um palíndromo')

else:
  print(f'A palavra não é um palíndromo')
```

```python
notas_globais = []

def media_ponderada():
  dados_para_calculo = []
  for i in range(3):
    try:

      nota = float(input(f'Digite a {i+1}ª nota: '))
      peso = float(input(f'Digite o peso para a {i+1}ª nota: '))
      dados_para_calculo.append({'nota': nota, 'peso': peso})
    except ValueError:
      print("Entrada inválida. Por favor, digite um número para notas e pesos.")
      return "Cálculo abortado devido a entrada inválida."

  total_ponderado = 0
  soma_dos_pesos = 0

  if not dados_para_calculo:
    return "Nenhum dado foi coletado para o cálculo da média ponderada."

  for item in dados_para_calculo:
    total_ponderado += item['nota'] * item['peso']
    soma_dos_pesos += item['peso']

  if soma_dos_pesos == 0:
    return "Não é possível calcular a média ponderada com peso total zero."
  else:
    return total_ponderado / soma_dos_pesos

resultado_media = media_ponderada()
print(f'A média ponderada é: {resultado_media}')
```

```python
palavra_input = str(input('Digite uma palavra: '))

def verificar_vogal(caractere):
  if caractere.lower() in 'aeiou':
    return True
  else:
    return False

def contar_vogais(palavra):
  count = 0
  for char in palavra:
    if verificar_vogal(char):
      count += 1
  return count

quantidade_vogais = contar_vogais(palavra_input)
print(f'A palavra "{palavra_input}" tem {quantidade_vogais} vogal(is).')
```

```python
raio = float(input('Digite o raio do círculo: '))

pi = 3.14

def calcular_circulo(raio):
  area = pi * (raio ** 2)
  return area

print(f'A área do círculo é: {calcular_circulo(raio)}')
```
